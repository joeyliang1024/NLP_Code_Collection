```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.common.exceptions import WebDriverException as wde
from tqdm.auto import tqdm
import pandas as pd
import time
```

```python
# use selenium to get the content of each chapter
chrome_options = Options()
# chrome_options.add_argument("--headless")
chrome_options.add_argument("--incognito") 
chromedriver = "/usr/local/bin/chromedriver"
driver = webdriver.Chrome(chromedriver, options=chrome_options)
driver.get("https://www.peoplemedia.tw/list/%E5%8F%B0%E8%AA%9E%E4%B8%96%E7%95%8C") # 主網頁
```

# Get all page url links

```python
to_be_craw_link = []
driver.get("https://www.peoplemedia.tw/list/%E5%8F%B0%E8%AA%9E%E4%B8%96%E7%95%8C") # 主網頁
time.sleep(5)
stop_url = 'https://www.peoplemedia.tw/list/%E5%8F%B0%E8%AA%9E%E4%B8%96%E7%95%8C#page-29'
while stop_url != driver.current_url:
    table = driver.find_element_by_id("area_list")
    candidate_link = table.find_elements_by_tag_name('a')
    to_be_craw_link += [elem.get_attribute('href') for elem in candidate_link]
    # click button 搜尋
    driver.find_element_by_class_name("page-link.next").click()
    time.sleep(5)
```

```python
with open("台語世界連結.txt", 'w') as f:
    for i in to_be_craw_link:
        f.write(i+"\n")
```

# get intext in url

```python
file = open('網站連結/台語世界連結.txt', 'r')
to_be_craw_link = file.readlines()
```

```python
crawed_text = []
url_link = []
titles = []
for url in tqdm(to_be_craw_link):
    try:
        driver.get(url) 
    except:
        driver.close()
        driver = webdriver.Chrome(chromedriver, options=chrome_options)
        driver.get(url)
    texts = [element.text for element in driver.find_elements_by_class_name('news_font2')]
    crawed_text+=texts
    title = driver.find_element_by_class_name('news_title').text
    titles+=[title for i in range(len(texts))]
    url_link+=[url for i in range(len(texts))]
    print("{} {} {} {}".foramt(title, len(crawed_text), len(url_link, len(titles))))
df = pd.DataFrame({
    '標題': titles,
    '內文': crawed_text,
    '連結': url_link
})
df.to_csv('台語世界.csv', index = False)
driver.close()
```

# 區分白話字還是漢羅
1. 可能可以用re檢查中文字 "\u4e00-\u9fff"
2. 格式難以對齊，無法做平行語料

```python
import re
```

```python
df = pd.read_csv("台語世界.csv")
df = df[df['內文'].isnull()==False]
df.to_csv("台語世界無空值.csv",index = False)
```

```python
courpus = []
sentence = []
add_han = True
add_white = True
df = pd.read_csv("台語世界無空值.csv")
title = df['標題'].iloc[0]
for i in range(df.shape[0]):
    #check add or not
    if str(df["內文"].iloc[i]) == "台灣字（白話字）版":
        #之後不是漢羅
        print(str(df["內文"].iloc[i]))
        add_han = False
    if str(df["內文"].iloc[i]) == "讀hōo你聽：":
        #之後不是白話字
        print(str(df["內文"].iloc[i]))
        add_white = False
    #check title
    if title != df['標題'].iloc[i]:
        courpus.append(sentence)
        title = df['標題'].iloc[i]
        #歸零
        sentence = []
        add_han = True
        add_white = True
    #add intext to sentence
    if len(re.findall("[\u4e00-\u9fff]",str(df["內文"].iloc[i])))!=0 and add_han == True:
        sentence.append((title, df["內文"].iloc[i], "漢羅"))
    elif len(re.findall("[\u4e00-\u9fff]",str(df["內文"].iloc[i])))==0 and add_white == True:
        sentence.append((title, df["內文"].iloc[i], "白話字"))
```

```python
han_lao = []
white = []
for sentences in courpus:
    han = ""
    whe = ""
    for sen in sentences:
        if sen[2] == "漢羅":
            han = han + sen[1]
        if sen[2] == "白話字":
            whe = whe + sen[1]
    han_lao.append(han)
    white.append(whe)
df1 = pd.DataFrame({
    '漢羅': han_lao,
    '白話字': white
})
df1.to_csv("台語世界平行語料.csv", index = False)
```

```python
with open("台語世界(漢羅).txt", 'w') as f:
    for i in han_lao:
        f.write(i+"\n\n")
with open("台語世界(白話字).txt", 'w') as f:
    for i in white:
        f.write(i+"\n\n")
```
